## General Framework

Each project must contain the 5 applications listed below. 

- **Welcome Application**: introduces the experiment, explains how the final payoff is determined (cumulative payoff of the main task and the control task), and collects consent. 
- **Main Task**: include the instructions.
- **Control Task**: a specific behavioral or cognitive task.
- **Demographics**: standard questionnaire.
- **Final Application**: thanks the participants and displays the final payoff.

##  Evaluation Criteria

- **Functionality**: Does the app run without errors, and are the tasks properly implemented?
- **Creativity**: Have the students added interesting features or enhancements (e.g., detailed instructions, feedback, etc.)?
- **Clarity**: Are the instructions and the flow of the experiment clear and logical?
- **Robustness**: Is the app resilient to user errors or unexpected inputs?
- **Data Collection**: Are the key variables recorded correctly for analysis?
